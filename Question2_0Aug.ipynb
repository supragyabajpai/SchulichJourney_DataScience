```python
import pandas as pd
import sqlalchemy as sa
import requests
```

```python
resource_id = '8a89caa9-511c-4568-af89-7f2174b4378c'
limit = 1000
offset = 0
```

```python
api_url = 'https://data.ontario.ca/api/3/action/datastore_search?resource_id={}&limit={}&offset={}'.format(resource_id,limit,offset)

print(api_url)
```

```python
api_response=requests.get(api_url)
api_response
```

```python

data=api_response.json()
data
```

```python
data['result']['records']
```

```python
covid_vaccine=pd.DataFrame(data['result']['records'])
covid_vaccine.tail()
```

```python

df['report_date'] = pd.to_datetime(df['report_date'])  
df['total_individuals_at_least_one'] = df['total_individuals_at_least_one'].astype(int)
df['total_individuals_fully_vaccinated'] = df['total_individuals_fully_vaccinated'].astype(int)
df['total_individuals_3doses'] = df['total_individuals_3doses'].astype(float)

df['new_doses'] = df['total_doses_administered'].diff()
df['new_doses_unvaccinated'] = df['total_individuals_at_least_one'].diff()
df['new_doses_one_vaccine'] = (df['total_individuals_at_least_one'] - df['total_individuals_fully_vaccinated']).diff()
df['new_doses_two_vaccines'] = (df['total_individuals_fully_vaccinated'] - df['total_individuals_3doses']).diff()
df['new_doses_three_vaccines'] = df['total_individuals_3doses'].diff()


```
